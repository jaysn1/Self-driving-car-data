```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
```

```
n_pts = 500
centers = [[-1,1], [-1, -1], [1, -1]]
X, y = datasets.make_blobs(n_samples=n_pts, centers=centers,random_state=123, shuffle='true', cluster_std=0.4)
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.scatter(X[y==2,0], X[y==2,1])
```

```
y_cat = to_categorical(y, num_classes=3)
```

```
model = Sequential()
model.add(Dense(3, input_shape=(2,), activation='softmax'))
model.compile(Adam(lr=0.1), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
h = model.fit(x=X, y=y_cat, batch_size=50, epochs=100, verbose=1)
```

```
plt.plot(h.history['loss'])
```
